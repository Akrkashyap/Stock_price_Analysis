```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Data/stock_data.csv")

df.head()
df.isnull().sum()
df.drop_duplicates(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])

df.set_index(
    'Date',
    inplace=True
)
plt.figure(figsize=(10,5))

plt.plot(
    df.index,
    df['Close']
)

plt.title(
    "Stock Closing Price Trend"
)

plt.xlabel(
    "Date"
)

plt.ylabel(
    "Closing Price"
)

plt.show()
df['MA30'] = (
    df['Close']
    .rolling(30)
    .mean()
)
plt.figure(figsize=(10,5))


plt.plot(
    df['Close'],
    label="Closing Price"
)


plt.plot(
    df['MA30'],
    label="30 Day Moving Average"
)


plt.title(
    "Stock Price With Moving Average"
)


plt.legend()

plt.show()
df['Daily_Return'] = (
    df['Close']
    .pct_change()
)
plt.figure(figsize=(10,5))


plt.plot(
    df['Daily_Return']
)


plt.title(
    "Daily Stock Returns"
)


plt.xlabel(
    "Date"
)

plt.ylabel(
    "Return"
)


plt.show()
plt.figure(figsize=(10,5))


plt.bar(
    df.index,
    df['Volume']
)


plt.title(
    "Trading Volume Analysis"
)


plt.show()
df.describe()
```
